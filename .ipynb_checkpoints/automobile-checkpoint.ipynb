## 3.0 Processing

Here's the code to load the data and find the percentage of NANs for the column 'normalized-losses.'

```python
#aim is to get a list of codes
#step 1 - import required modules
import pandas as pd
import unittest
import numpy as np
from IPython.display import Markdown as md
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook

#step - create a functions that bins continous variables of columns into 3 bins

def bin_continous_var(df,col, n):
    
    df["cut"+col]= pd.cut(df[col], n)
    

#step - read the file called automobile data

automobile_file = pd.read_csv("automobile.data.csv")

#step - replace missing or unknown values, represented with ? to NAN, and change the dataframe by using NAN
automobile_file.replace("?", np.nan, inplace=True)

#step - find the percentage of NAN of the column 'normalized-losses'
na_mean=automobile_file['normalized-losses'].isna().mean()*100

#step - show the percentage of missing or NAN of the column normalized-losses.
md("{}% represents the percentage of missing or NAN of the column normalised-losses".format(na_mean))

#step - find the subset not null for the column 'normalized-losses'
droped_nan = automobile_file[automobile_file['normalized-losses'].notnull()]

#step - change data type of the column 'normalized-losses' to float because this is needed for the violin plot latter
droped_nan.loc[:,['normalized-losses']]=droped_nan['normalized-losses'].astype('float64')
```

```python
#step - create a wide figure
fig=plt.figure(figsize=(10,6))

#step - create a violin plot and get the axis object
chart=sns.violinplot('make', y='normalized-losses', data=droped_nan)

#step - rotate the x-axis labels 45 degrees and make room for the x-axis labels below.
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)

fig.subplots_adjust(bottom=0.2)
```
